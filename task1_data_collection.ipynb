```python
import requests
import json
import time
from datetime import datetime
import os


def get_top_story_ids():
    url = "https://hacker-news.firebaseio.com/v0/topstories.json"
    try:
        response = requests.get(url)
        print("status code:", response.status_code)
        print("Response type:", type(response))
        if response.status_code == 200:
            return response.json()[:500]
        else:
            print("Error fetching IDs")
            return []
    except Exception as e:
        print("Exception", e)
        return []


def get_story_details(story_id):
    url = f"https://hacker-news.firebaseio.com/v0/item/{story_id}.json"
    try:
        response = requests.get(url)
        if response.status_code == 200:
            return response.json()
    except:
        return None


def categorize(title):
    title = title.lower()

    if any(word in title for word in ["ai", "tech", "software", "api", "gpu"]):
        return "technology"
    elif any(word in title for word in ["war", "government", "election"]):
        return "worldnews"
    elif any(word in title for word in ["nfl", "fifa", "team", "player"]):
        return "sports"
    elif any(word in title for word in ["physics", "biology", "nasa"]):
        return "science"
    elif any(word in title for word in ["movie", "music", "netflix"]):
        return "entertainment"
    else:
        return "other"


def main():
    ids = get_top_story_ids()
    results = []
    stories = []  # Initialize the 'stories' list here

    for i in ids:
        data = get_story_details(i)

        if data and "title" in data:
            category = categorize(data["title"])

            results.append({
                "post_id": data["id"],
                "title": data["title"],
                "category": category
            })

    time.sleep(2)

    today = datetime.now().strftime("%Y-%m-%d")
    file_name = f"data/trends_{today}.json"

    # Ensure the 'data' directory exists
    os.makedirs("data", exist_ok=True)

    with open(file_name, "w") as f:
        json.dump(results, f, indent=4)


    print (f"\nTotal stories collected: {len(results)}")
    print(f"saved to {file_name}")

if __name__ == "__main__":
    main()
```
